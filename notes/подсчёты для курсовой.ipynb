## Изучение частотности вариантов предлогов

```python
import re
from itertools import combinations

def variants(wparen):
    x = re.split(r'(?=[(])|(?<=[)])', wparen)
    ix = [i for i,s in enumerate(x) if "(" in s]
    v = [()]
    for i in range(1, len(ix) + 1):
        v.extend([z for z in combinations(ix, i)])
    return v, x

def vstrings(wparen):
    v, x = variants(wparen)
    s = []
    for vrs in v:
        sx = ""
        for i, c in enumerate(x):
            cx = c.strip("()")
            if i not in vrs:
                sx += cx
        s.append(sx)
    return s

def syllables(s):
    return len(re.findall(r'[ёыеауоэяию]', s))
    
```

```python
[(syllables(W), W) for W in vstrings("в(о)")]
```

```python
def for_property(prop, variants):
    for v in variants:
        try:
            return dataset.Dataset(f"НКРЯ_{v}_ВАР_{prop}", "..")
        except FileNotFoundError:
            pass
    return None
```

```python
def variants_freq(prop, rule):
    vs = vstrings(rule)
    freq = {}
    for v in vs:
        freq[v] = 0
    d = for_property(prop, vs)
    if d is None:
        return freq
    for item in d.j["nk:datasetContent"]["items"]:
        sel = [subitem for subitem in item["text"] if subitem["status"] == "selected"]
        for sbi in sel:
            for v in vs:
                if sbi["text"] == v:
                    freq[v] += 1
    return freq
    
```

```python
S = 'без(о), в(о), для, до, за, из(о), из-за, из-под(о), к(о), на, над(о), о(б)(о), от(о), п(е)ред(о), по, по-за, по-над(о), по-под(о), под(о), при, про, с(о), у, ч(е)рез(о)'
V = S.split(", ")
P = ["АмальгамыИнтерпозиция", "АппроксИнверсия", "ДвРефлексивИнтерпозиция", "ОтрМестИнтерпозиция", "Удвоение"]
T = []
for v in V:
    for p in P:
        vf = variants_freq(p, v)
        vf_k = list(vf.keys())
        vf_k = sorted(vf_k, key=lambda k: vf[k], reverse=True)
        if syllables(vf_k[0]) != 0:
            vf = {vf_k[0]: vf[vf_k[0]]}
        for (word, freq) in vf.items():
            T.append((len(word.replace("-", "")), freq))
print(T)
l1 = [x[0] for x in T]
l2 = [x[1] for x in T]
print(l1)
print(l2)

#variants_freq("АмальгамыИнтерпозиция", "под(о)")
```
